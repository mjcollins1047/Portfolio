```python
# Import libraries

import pandas as pd
```

```python
# Load confirmed cases covid data for the US

df_confirmed = pd.read_csv('time_series_covid19_confirmed_US.csv')
```

```python
df_confirmed.shape
```

```python
# check the confirmed dataframe

df_confirmed.head()
```

```python
# Transform the datafram from wide to long format

confirmed = pd.melt(df_confirmed, id_vars = df_confirmed.columns[:11],
                   value_vars = df_confirmed.columns[11:],
                   var_name = 'date',
                   value_name = 'confirmed')

# View the new dataframe

confirmed.tail()
```

```python
# Load the confirmed covid death data for the US

df_death = pd.read_csv('time_series_covid19_deaths_US.csv')
```

```python
df_death.shape
```

```python
# check the deaths dataframe

df_death.head()
```

```python
# Transform the datafram from wide to long format

death = pd.melt(df_death, id_vars = df_death.columns[:12],
                   value_vars = df_death.columns[12:],
                   var_name = 'date',
                   value_name = 'deaths')

# View the new dataframe

death.head()
```

```python
# Merge the tables

data = confirmed.merge(death, how = 'inner', on = ['UID', 'iso2', 'iso3', 'code3', 'FIPS',
                                                   'Admin2', 'Province_State', 'Country_Region',
                                                   'Lat', 'Long_', 'Combined_Key', 'date'])
```

```python
# View the merged data
data.head()
```

```python
list(data)
```

```python
# Create a csv file to use in PowerBI

data.to_csv('covid_us_death_confirmed_data.csv')
```
